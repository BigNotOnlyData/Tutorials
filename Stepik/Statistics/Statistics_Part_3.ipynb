<span style="color:orange; font-weight: bold; font-size:12pt">–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –≤–∑–¥—É—Ç–∏—è –¥–∏—Å–ø–µ—Ä—Å–∏–∏ (VIF)</span> - –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ —Å–∫–æ–ª—å–∫–æ —Ö–æ—Ä–æ—à–æ –≤—ã–±—Ä–∞–Ω–Ω—ã–π –ø—Ä–µ–¥–∏–∫—Ç–æ—Ä –æ–±—ä—è—Å–Ω—è–µ—Ç—Å—è –¥—Ä—É–≥–∏–º–∏ –ø—Ä–µ–¥–∏–∫—Ç–æ—Ä–∞–º–∏.

$$VIF = \frac{1}{1 - R^{2}_{j}}$$
–ì–¥–µ $R^{2}_{j}$ - –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–µ—Ç–µ—Ä–º–∏–Ω–∞—Ü–∏–∏ j-–≥–æ –ø—Ä–µ–¥–∏–∫—Ç–æ—Ä–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö.

<span style="color:#00d01b; font-weight: bold; font-size:12pt">–ß—Ç–æ –¥–µ–ª–∞—Ç—å?</span>

+ –ü—Ä–∏–º–µ–Ω–∏—Ç—å –º–µ—Ç–æ–¥ –≥–ª–∞–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç –∏–ª–∏ —Ñ–∞–∫—Ç–æ—Ä–Ω—ã–π –∞–Ω–∞–ª–∏–∑ (—á—Ç–æ–±—ã –æ–±—ä–µ–¥–∏–Ω–∏—Ç—å —Å–∏–ª—å–Ω–æ –∫–æ—Ä—Ä–µ–ª–∏—Ä—É—é—â–∏–µ –ø—Ä–µ–¥–∏–∫—Ç–æ—Ä—ã –≤ –æ–¥–∏–Ω –Ω–æ–≤—ã–π –ø—Ä–µ–¥–∏–∫—Ç–æ—Ä) –∏ –∑–∞–Ω–æ–≤–æ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –º–æ–¥–µ–ª—å

![–î–∏—Å–ø–µ—Ä—Å–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤](img/Vif.png)



```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from patsy import dmatrices

%matplotlib inline

url = 'https://gist.githubusercontent.com/christophsax/178d34245afdd6e187b1fff72dbe7448/raw/f5f4189f949f117bee4e82e4aa75c104ed20b4f4/swiss.csv'
data = pd.read_csv(url)
data.rename(columns={'Infant.Mortality': 'Mortality'}, inplace=True)


# —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–æ –º–∞—Ç—Ä–∏—Ü–µ —Ä–µ–≥—Ä–µ—Å–∏–∏
axes = pd.plotting.scatter_matrix(data, figsize=(10,10), diagonal='kde', grid=True)
corr = data.corr()

# —ç—Ç–æ –¥–ª—è –∫—Ä–∞—Å—Å–∏–≤–æ–π –∞–Ω–Ω–æ—Ç–∞—Ü–∏–∏
for i, j in zip(*plt.np.triu_indices_from(axes, k=1)):
    value = corr.values[i,j]
    if abs(value) <= 0.25:
        fontsize = 14
    else:
        fontsize = round(56*abs(value))
    axes[i, j].annotate(
        "%.2f" %value, (0.5, 0.5), xycoords='axes fraction'
        , ha='center', va='center', fontsize=fontsize
    )
plt.show()

print(corr)


# —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º VIF
predictors = " + ".join(list(
    set(data.columns) - set(['Fertility', 'Location'])
    ))
y, X = dmatrices('Fertility ~' + predictors, data, return_type='dataframe')


vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
print(f'\nVIF:\n{vif.T}')
```

(–ü—Ä–∏–º–µ—Ä: –∏–∑–º–µ—Ä–µ–Ω–∏–π —Å –ø–ª–∞—Ü–µ–±–æ –±–æ–ª—å—à–µ, —á–µ–º —Å –ª–µ–∫–∞—Ä—Å—Ç–≤–æ–º):
  * –ò—Å–∫–∞–∂–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
+ <span style="color:orange; font-weight: bold; font-size:12pt">–ö–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö:</span>

| *–ü—Ä–∏–º–µ—Ä –ø—Å–µ–≤–¥–æ—Ä–µ–ø–ª–∏–∫–∞—Ü–∏–∏ (–ï—Å–ª–∏ –ø—Ä–Ω–∏–Ω—è—Ç—å —á—Ç–æ –∏–∑–º–µ—Ä–µ–Ω–∏—è –ù–ï–∑–∞–≤–∏–º—ã–µ) –∏ —Å–Ω–∏–∂–µ–Ω–∏—è —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Ç–µ—Å—Ç–∞ (–ï—Å–ª–∏ –ø—Ä–Ω–∏–Ω—è—Ç—å —á—Ç–æ –∏–∑–º–µ—Ä–µ–Ω–∏—è –∑–∞–≤–∏–º—ã–µ)* |

<span style="color:orange; font-weight: bold; font-size:12pt">–≠—Ñ—Ñ–µ–∫—Ç</span> - –≤–ª–∏—è–Ω–∏–µ –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π, —Å –ø–æ–º–æ—â—å—é –∫–æ—Ç–æ—Ä–æ–π –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–µ–º –∑–∞–≤–∏—Å–∏–º—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é.

<br>

<span style="color:orange; font-weight: bold; font-size:12pt">Jackknife</span> - –º–µ—Ç–æ–¥ –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã–π –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç–∏ –≤ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–º –≤—ã–≤–æ–¥–µ –∏ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–∏—è –æ—Ü–µ–Ω–∫–∏, –µ—Å–ª–∏ –æ–Ω–∞ c–º–µ—â–µ–Ω–Ω–∞—è.

<span style="color:#00d01b; font-weight: bold; font-size:12pt">–¢–µ–æ—Ä–∏—è:</span>

$X$ - —Å–ª—É—á–∞–π–Ω–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è (–Ω–∞–ø—Ä–∏–º–µ—Ä —Ä–æ—Å—Ç);

$\left[X_{1}, ..., X_{n} \right]$ - –≤—ã–±–æ—Ä–∫–∞ –∏–∑ n-—ç–ª–µ–º–µ–Ω—Ç–æ–≤;

$\theta$ - –ø–∞—Ä–∞–º–µ—Ç—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã (–Ω–∞–ø—Ä–∏–º–µ—Ä —Å—Ä–µ–¥–Ω–µ–µ, –¥–∏—Å–ø–µ—Ä—Å–∏—è, –º–∞–∫—Å–∏–º—É–º –∏ —Ç.–¥) –≤ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏;

$\hat{\theta}$ - –æ—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ (—Å—Ä–µ–¥–Ω–µ–µ, –¥–∏—Å–ø–µ—Ä—Å–∏—è, –º–∞–∫—Å–∏–º—É–º –∏ —Ç.–¥) –≤ –≤—ã–±–æ—Ä–∫–µ;

–í –∏–¥–µ–∞–ª—å–Ω–æ–º —Å–ª—É—á–∞–µ **–º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ** –æ—Ü–µ–Ω–∫–∏ –≤ –≤—ã–±–æ—Ä–∫–µ —Ä–∞–≤–Ω–æ –ø–∞—Ä–∞–º–µ—Ç—Ä—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏: $$E(\hat{\theta})=\theta$$

–ù–æ –æ—Ü–µ–Ω–∫–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å **—Å–º–µ—â–µ–Ω–∞**:
$$E(\hat{\theta})=\theta + \frac{\alpha }{n} + \frac{\beta }{n^{2}} +  \frac{\gamma}{n^{3}} + ...$$

–ì–¥–µ $\frac{\alpha }{n} + \frac{\beta }{n^{2}} +  \frac{\gamma}{n^{3}} + ...$ - —Å–æ—Å—Ç–∞–≤–ª—è—é—â–∏–µ –æ—à–∏–±–∫–∏.

**–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ** –ø–æ –≤—ã–±–æ—Ä–∫–∞–º –±–µ–∑ –æ–¥–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞:

$$\hat{\theta}_{(.)} = \frac{\sum_{i=1}^{n}{{\hat\theta_{(i)} }}}{n}$$

–ì–¥–µ $\hat\theta_{(i)}$ - –æ—Ü–µ–Ω–∫–∞ –Ω–∞ –≤—ã–±–æ—Ä–∫–µ –±–µ–∑  —ç–ª–µ–º–µ–Ω—Ç–∞ **i**.
![–°–º–µ—à–∞–Ω–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è](img/jackknife.png)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def bad_var_estimator(x):
    """–ò—Å–ø–æ—Ä—á–µ–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–∏—Å–ø–µ—Ä—Å–∏–∏"""
    n = len(x)
    return np.var(x)*(n-1)/n

def JN_bias_correction(x, estimator):
    """–§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç: 
    –º–∞—Å—Å–∏–≤ - x: numpy.ndarray;
    –∞–≥—Ä–µ–≥–∏—Ä—É—é—â–∞—è —Ñ—É–Ω–∫—Ü–∏—è - estimator (—Å—Ä–µ–¥–Ω–µ–µ, –¥–∏—Å–ø–µ—Ä—Å–∏—è, –º–∞–∫—Å–∏–º—É–º –∏ —Ç.–¥).
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω–Ω—É—é –æ—Ü–µ–Ω–∫—É –ø–æ –º–µ—Ç–æ–¥—É Jackknife"""
    
    # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ
    n = len(x)
    
    # –ó–∞–≥–æ—Ç–æ–≤–∫–∞ –º–∞—Å—Å–∏–≤–∞ –æ—Ü–µ–Ω–æ–∫ –±–µ–∑ –æ–¥–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞
    theta_stars = np.zeros_like(x, dtype='float')
    
    # –ó–∞–ø–æ–ª–Ω—è–µ–º –º–∞—Å—Å–∏–≤ (theta_stars) –æ—Ü–µ–Ω–∫–∞–º–∏ –ø–æ –≤—ã–±–æ—Ä–∫–∞–º  –±–µ–∑ –æ–¥–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞
    for i in range(n):
        sample = np.concatenate([x[:i], x[i+1:]], axis=0)
        theta_stars[i] = estimator(sample)
    
    # –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–∞—è –æ—Ü–µ–Ω–∫–∞
    theta_hat = estimator(x)
    
    # —Å—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞ –ø–æ –≤—Å–µ–º –≤—ã–±–æ—Ä–∫–∞–º
    theta_dot = np.mean(theta_stars)
    
    # –≤—ã—á–∏—Å–ª—è–µ–º —Å–º–µ—â–µ–Ω–∏–µ
    bias_jack = (theta_dot - theta_hat) * (n - 1)
    
    # –∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä—É–µ–º –æ—Ü–µ–Ω–∫—É
    theta_hat_jack = theta_hat - bias_jack
    
    return theta_hat_jack
```

```python
start = 3  # –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ
test = 100  # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–µ—Å—Ç–æ–≤—ã—Ö –≤—ã–±–æ—Ä–æ–∫ –æ–¥–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞
sample_sizes = np.arange(start, 51)  # –º–∞—Å—Å–∏–≤ —Å —Ä–∞–∑–º–µ—Ä–∞–º–∏ –≤—ã–±–æ—Ä–æ–∫

# –ó–∞–≥–æ—Ç–∞–≤–ª–∏–≤–∞–µ–º –º–∞—Å—Å–∏–≤—ã –ø–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å —Ç–∏–ø–æ–º float –∏–Ω–∞—á–µ –∑–Ω–∞—á–µ–Ω–∏—è –±—É–¥—É—Ç –æ–∫—Ä—É–≥–ª—è—Ç—Å—è –¥–æ int
results_good = np.zeros_like(sample_sizes, dtype='float')
results_bad = np.zeros_like(sample_sizes, dtype='float')
results_corrected = np.zeros_like(sample_sizes, dtype='float')

for n in sample_sizes:
    # –≥–µ–Ω–µ—Ä–∏–º test-–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑ (—Å—Ç–æ–ª–±—Ü—ã) –≤—ã–±–æ—Ä–∫–∏ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å—é n (—Å—Ç—Ä–æ–∫–∏)
    samples = np.random.rand(n, test)
    
    # –ø—Ä–∏–º–µ–Ω—è–µ–º –∫ —Å—Ç–æ–ª–±—Ü–∞–º —Ñ—É–Ω–∫—Ü–∏—é-–æ—Ü–µ–Ω—â–∏–∫
    good_estimations = np.apply_along_axis(np.var, 0, samples)
    bad_estimations = np.apply_along_axis(bad_var_estimator, 0, samples)
    corrected_estimations = np.apply_along_axis(JN_bias_correction, 0, samples, estimator=bad_var_estimator) 
    
    # —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Å—Ç—Ä–µ–¥–Ω–µ–µ –∏ –∑–∞–ø–∏—Å—ã–≤–∞–µ–º –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
    results_good[n-start] = np.mean(good_estimations)
    results_bad[n-start] = np.mean(bad_estimations)
    results_corrected[n-start] = np.mean(corrected_estimations) 

# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫–∏    
plt.figure(figsize=(15, 5))
plt.scatter(sample_sizes, results_corrected, label='corrected', color='b')
plt.scatter(sample_sizes, results_bad, label='bad', color='r')
plt.scatter(sample_sizes, results_good, label='good', color='g')
plt.legend()
```

```python
# –ø—Ä–∏–º–µ—Ä —Å —Ñ—É–Ω–∫—Ü–∏–µ–π –º–∞–∫—Å–∏–º—É–º–∞
for n in sample_sizes:
    samples = np.random.rand(n, test)
    
    bad_estimations = np.apply_along_axis(np.max, 0, samples)
    corrected_estimations = np.apply_along_axis(JN_bias_correction, 0, samples, estimator=np.max) 
    
    results_bad[n-start] = np.mean(bad_estimations)
    results_corrected[n-start] = np.mean(corrected_estimations) 

plt.figure(figsize=(15, 5))
plt.plot(sample_sizes, results_corrected, label='corrected', color='b')
plt.plot(sample_sizes, results_bad, label='bad', color='r')
plt.legend()
```

+ <span style="color:orange; font-weight: bold; font-size:12pt">–ü–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å–Ω—ã–π</span>:
    - –ì–µ–Ω–µ—Ä–∏–º –±—É—Ç—Å—Ä–∞–ø—ç–≤—Å–∫–∏–µ –≤—ã–±–æ—Ä–∫–∏ –∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ–º –∏—Å—Å–ª–µ–¥—É–µ–º—É—é —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –¥–ª—è –∫–∞–∂–¥–æ–π —Ç–∞–∫–æ–π –≤—ã–±–æ—Ä–∫–∏. –î–∞–ª–µ–µ —Å—Ç—Ä–æ–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫ –∏ –Ω–∞ –ø–æ–ª—É—á–µ–Ω–Ω–æ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ç—Ä–æ–∏–º –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª.
    - –ü–ª–æ—Ö–æ —Ä–∞–±–æ—Ç–∞–µ—Ç –µ—Å–ª–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –Ω–µ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ–µ.
    
+ <span style="color:orange; font-weight: bold; font-size:12pt">–ë–∞–∑–æ–≤—ã–π (—Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã–π)</span> - –ø—Ä–µ–¥–ø–æ—á—Ç–∏—Ç–µ–ª—å–Ω–µ–µ.
   - –ò–¥–µ—è –≤ —Ç–æ–º —á—Ç–æ—é—ã –ø—Ä–∏–±–ª–∏–∑–∏—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è: $\mu -\bar{x} \approx \bar{x} - \bar{x}^{*}$. –ì–¥–µ $\mu$ - —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –≤ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏, $\bar{x}$ - —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –≤ –≤—ã–±–æ—Ä–∫–µ, $\bar{x}^{*}$ - —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –≤ –±—É—Ç—Å—Ä–∞–ø–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ.
   - –ù–∞–ø—Ä–∏–º–µ—Ä 90% –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $\mu -\bar{x}: P(\mu \in [\delta _{0.05} + \bar{x}, \delta _{0.95} + \bar{x}]) = 0.9$. –ì–¥–µ $\delta _{0.05}$ –∏ $\delta _{0.95}$ - 5% –∏ 95% –∫–≤–∞–Ω—Ç–∏–ª–∏ –±–∞–∑–æ–≤–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $\bar{x} - \bar{x}^{*}$.
   
üí° **–ü–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å–Ω—ã–π –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª** –±–µ—Ä—ë—Ç—Å—è "–≤ –ª–æ–±" –∏–∑ –∫–≤–∞–Ω—Ç–∏–ª–µ–π –±—É—Ç—Å—Ç—Ä–∞–ø–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ (–≤ —Å–ª—É—á–∞–µ —Å–æ —Å—Ä–µ–¥–Ω–∏–º ‚Äì –∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $\bar{x}^{*}$

üí° **–ë–∞–∑–æ–≤—ã–π –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª**, –≤ —Å–≤–æ—é –æ—á–µ—Ä–µ–¥—å, ‚Äì –Ω–∞ –æ—Å–Ω–æ–≤–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $\bar{x} - \bar{x}^{*}$ (–∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $\mu -\bar{x}$).

```python
np.random.seed(42)

# –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ
x_bar = 10  

# –º–∞—Å—Å–∏–≤ —Å—Ä–µ–¥–Ω–∏—Ö –ø–æ –±—É—Ç—Å—Ç—Ä—ç–ø–æ–≤—Å–∫–∏–º –≤—ã–±–æ—Ä–∫–∞–º
x_barstars = np.random.normal(loc = x_bar, scale = 1.0, size = 700)

# –∫–≤–∞–Ω—Ç–∏–ª–∏ 5% –∏ 95% –Ω–∞ –±—É—Ç—Å—Ä—ç–ø–æ–≤—Å–∫–æ–º (–ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å–Ω–æ–º) —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –±—É—Ç—Å—Ä—ç–ø–æ–≤—Å–∫–∏—Ö —Å—Ä–µ–¥–Ω–∏—Ö
low = np.quantile(x_barstars, 0.05)
high = np.quantile(x_barstars, 0.95)

# –∫–≤–∞–Ω—Ç–∏–ª–∏ 5% –∏ 95% –Ω–∞ –±–∞–∑–æ–≤–æ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ x_bar - x_barstars
low_basic = np.quantile(x_bar - x_barstars, 0.05)
high_basic = np.quantile(x_bar - x_barstars, 0.95)

# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫–∏
plt.figure(figsize=(15, 5))
plt.hist(x_barstars, bins=100)
plt.hist(x_bar - x_barstars, bins=100)
plt.vlines(low, ymin=0, ymax=30, color='r')
plt.vlines(high, ymin=0, ymax=30, color='r')
plt.vlines(low_basic, ymin=0, ymax=30, color='g')
plt.vlines(high_basic, ymin=0, ymax=30, color='g')
plt.vlines(-((high - low) / 2 + (high + low) / 2 - x_bar), ymin=25, \
           ymax=35, color='black', linestyles='--')
plt.show()
```

```python
print('–£–ø—Ä–æ—â–µ–Ω–Ω–∞—è —Ñ–æ—Ä–º—É–ª–∞:', -((high - low) / 2 + (high + low) / 2 - x_bar) == x_bar - high)
# –±–∞–∑–æ–≤–π 5% –∫–≤–∞–Ω—Ç–∏–ª—å  –≤—ã—Ä–∞–∂–µ–Ω–Ω—ã–π —á–µ—Ä–µ–∑ –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å–Ω—ã–π 95% –∫–≤–∞–Ω—Ç–∏–ª—å
low_basic_exp = x_bar - high
print('–†–∞–≤–µ–Ω—Å—Ç–≤–æ –∫–≤–∞–Ω—Ç–∏–ª–µ–π –ø–æ —Ñ–æ—Ä–º—É–ª–µ –∏ –∏–∑ –ø–∞–∫–µ—Ç–∞ numpy:', np.isclose(low_basic_exp, low_basic))
```
